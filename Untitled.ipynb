```python
df = df.drop(columns=['id','poutcome','day','month'])
```

```python
df['job'] = df['job'].apply(lambda x:'management' if x == 'unknown' else x)
df['contact'] = df['contact'].apply(lambda x:'cellular' if x == 'unknown' else x)
```

```python
pt = PowerTransformer(method='yeo-johnson', standardize=True)
df['balance'] = pt.fit_transform(df[['balance']])
```

```python
numerical_cols = [c for c in df.columns if df[c].dtypes != 'O' and c!='y']
categorical_cols = [c for c in df.columns if df[c].dtypes == 'O' and c!='y']
```

```python
categorical_cols
```

```python
numerical_cols
```

```python
def numerical_features_plot(df, feature, target):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    sns.boxplot(data=df, x=feature, y=target, hue=target, orient='h', ax=axes[0])
    axes[0].set_title(f'Boxplot of {feature}')

    sns.violinplot(data=df, x=feature, y=target, hue=target, orient='h', ax=axes[1])
    axes[1].set_title(f'Violinplot of {feature}')

    sns.histplot(data=df, x=feature, hue=target, kde=True, ax=axes[2])
    axes[2].set_title(f'Distribution of {feature}')
    plt.show()
```

```python
for c in numerical_cols:
    numerical_features_plot(df,c,'y')
```

```python
for c in categorical_cols:
    w = 20 if c == 'job' else 6
    plt.figure(figsize=(w,4))
    sns.countplot(data=df,x=c,hue='y')
    plt.show()
```

```python
df = pd.get_dummies(df,columns=categorical_cols)
df
```

```python
X = df.drop(columns=['y'])
y = df['y']
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
rfc = RandomForestClassifier(n_estimators=500, min_samples_split=10, min_samples_leaf=4,max_features='sqrt',max_depth=None, bootstrap=False)
rfc.fit(X_train,y_train)
y_pred = rfc.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```
